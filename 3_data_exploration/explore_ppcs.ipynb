### Interpretation

The heatmap shows strong positive correlations between the main mental health indicators. Psychological Distress Aggregate is highly correlated with Depression SD (r = 0.70) and PTSD Index (r = 0.64), indicating that individuals with higher depression or PTSD symptoms also experience higher general psychological distress. Distress SWIN SD also shows moderate correlations with both depression (r = 0.31) and general distress (r = 0.59). These relationships suggest overlapping dimensions of trauma-related mental health challenges within the population.
